```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('D:\Python Programs\Data Science\Wholesale customers data.csv')
data.head()
```

```python
from sklearn.preprocessing import normalize
data_scaled = normalize(data)
data_scaled = pd.DataFrame(data_scaled, columns=data.columns)
data_scaled.head()
```

```python
# from sklearn.preprocessing import StandardScaler
# scaled_data = StandardScaler().fit(data).transform(data)
# scaled_data = pd.DataFrame(scaled_data,columns = data.columns)
# scaled_data.head()
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))  
plt.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data_scaled,method = 'ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')  
cluster.fit_predict(data_scaled)
```

```python
cluster.labels_
```

```python
data_scaled['labels'] = cluster.labels_
```

```python
data_scaled.head()
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(data_scaled['Milk'], data_scaled['Grocery'], c = cluster.labels_,cmap = 'rainbow') 
```
